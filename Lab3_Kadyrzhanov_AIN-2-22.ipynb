```python
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def quadratic_function(x):
    return 2*x**2 - 3*x + 1  

x = np.linspace(-10, 10, 400)
y = quadratic_function(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = 2x^2 - 3x + 1")

plt.title("График квадратичной функции")
plt.xlabel("ОСЬ X")
plt.ylabel("ОСЬ Y")
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def harmonic_function(x):
    return np.sin(0.5 * x)

x = np.linspace(0, 4 * np.pi, 400)
y = harmonic_function(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = sin(0.5x)")

plt.title("График гармонической функции")
plt.xlabel("ОСЬ X")
plt.ylabel("ОСЬ Y")
plt.grid(True)

plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

axs[0].plot(x, quadratic_function(x), label="y = 2x^2 - 3x + 1")
axs[0].set_title('Квадратичная функция')
axs[0].set_xlabel('ОСЬ X')
axs[0].set_ylabel('ОСЬ Y')
axs[0].grid(True)

axs[1].plot(x, harmonic_function(x), label="y = sin(0.5x)")
axs[1].set_title('Гармоническая функция')
axs[1].set_xlabel('ОСЬ X')
axs[1].set_ylabel('ОСЬ Y')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.plot(x, quadratic_function(x), label="y = 2x^2 - 3x + 1")
plt.plot(x, harmonic_function(x), label="y = sin(0.5x)")

plt.title("Квадратичная и Гармоническая Функции")
plt.xlabel("ОСЬ X")
plt.ylabel("ОСЬ Y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
t = np.linspace(0, 4 * np.pi, 400)
x_param = np.cos(3 * t)
y_param = np.sin(2 * t)

plt.figure(figsize=(8, 6))
plt.plot(x_param, y_param)

plt.title("x(t) = cos(3t), y(t) = sin(2t)")
plt.xlabel("ОСЬ X")
plt.ylabel("ОСЬ Y")
plt.grid(True)
plt.show()
```

```python
years = np.array([2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022])
values = np.array([4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32])

plt.figure(figsize=(10,6))
plt.bar(years, values, color='skyblue')
plt.xlabel('Год')
plt.ylabel('Прожиточный минимум')
plt.title('(2012-2022)')
plt.xticks(years, rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,8))
plt.pie(values, labels=years, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('(2012-2022)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def simple_surface_function(x, y):
    return np.sin(x) + np.cos(y)

x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
x, y = np.meshgrid(x, y)

z = simple_surface_function(x, y)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(x, y, z)
plt.show()
```
